```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import rc
from pandas.plotting import register_matplotlib_converters

%matplotlib inline

register_matplotlib_converters()
sns.set(style='whitegrid', palette='muted', font_scale=1.5)

rcParams['figure.figsize'] = 22, 10

RANDOM_SEED = 42

np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)
pd.options.plotting.backend = 'plotly'
```

```python
# Reading the dataset and closing price
data = pd.read_csv('./nifty50-historical-data.csv',parse_dates=['Date'], index_col='Date')[['Price']]
```

```python
# Cleaning the price values and converting to float
data.Price = data.Price.str.replace(',','').astype(float)
```

```python
data.sort_values(by='Date',inplace=True)
```

```python
data.plot()
```

```python
# Split the data into train and test set
train_size = int(len(data) * 0.95)
test_size = len(data) - train_size
train, test = data.iloc[0:train_size], data.iloc[train_size:len(data)]
print(train.shape, test.shape)
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

train['Price'] = sc.fit_transform(train[['Price']])
test['Price'] = sc.fit_transform(test[['Price']])
```

```python
# Hepler method to generaate sequences from the dataset, that the LSTM Autoencoder will understand.
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)        
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
TIME_STEPS = 30

# reshape to [samples, time_steps, n_features]

X_train, y_train = create_dataset(train[['Price']], train.Price, TIME_STEPS)
X_test, y_test = create_dataset(test[['Price']], test.Price, TIME_STEPS)

print(X_train.shape)
```

## How to detect Anomaly using LSTM-Autoencoder

The broad idea is that if the predictive error of the LSTM Autoencoder is beyond a threshold, i.e., it cannot resconstruct this data(the primary task of any autoencoder), and that might be an anomaly(a rare event).

Essentially, we are giving input to our model 30 days of closing price of Nifty50, trying to predict the same using the Autoencoder architecture, and if the error exceeds a threshold, we can say it was an anomaly.

```python
# Designing the autoencoder
model = keras.Sequential()
model.add(keras.layers.LSTM(
    units=64, 
    input_shape=(X_train.shape[1], X_train.shape[2])
))
model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.RepeatVector(n=X_train.shape[1]))
model.add(keras.layers.LSTM(units=64, return_sequences=True))
model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.TimeDistributed(keras.layers.Dense(units=X_train.shape[2])))
model.compile(loss='mae', optimizer='adam')
```

```python
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.1,
    shuffle=False
)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend();
```

```python
# Predict the reconstructed output using the trained autoencoder model
X_train_pred = model.predict(X_train)

train_mae_loss = np.mean(np.abs(X_train_pred - X_train), axis=1)
```

```python

sns.distplot(train_mae_loss, bins=50, kde=True);
```

```python
X_test_pred = model.predict(X_test)

test_mae_loss = np.mean(np.abs(X_test_pred - X_test), axis=1)
```

```python
THRESHOLD = 0.65

test_score_df = pd.DataFrame(index=test[TIME_STEPS:].index)
test_score_df['loss'] = test_mae_loss
test_score_df['threshold'] = THRESHOLD
test_score_df['anomaly'] = test_score_df.loss > test_score_df.threshold
test_score_df['Price'] = test[TIME_STEPS:].Price
```

```python
plt.plot(test_score_df.index, test_score_df.loss, label='loss')
plt.plot(test_score_df.index, test_score_df.threshold, label='threshold')
plt.xticks(rotation=25)
plt.legend();
```

```python
anomalies = test_score_df[test_score_df.anomaly == True]
anomalies.head()
```

```python
plt.plot(
  test[TIME_STEPS:].index, 
  sc.inverse_transform(test[TIME_STEPS:].Price), 
  label='close price'
);

sns.scatterplot(
  anomalies.index,
  sc.inverse_transform(anomalies.Price),
  color=sns.color_palette()[3],
  s=52,
  label='anomaly'
)
plt.xticks(rotation=25)
plt.legend();
```
